# Analysis on Combined Datasets & Model Building and Evaluation

In this notebook, we combine all the datasets together using meaningful pipelines investigated in each individual dataset analysis notebooks. We then perform EDA on the combined dataset, visualizing interesting findings. Finally, we define the models, train them, and evaluate the results.

__What type of machine learning problem are you investigating?__

In this notebook, we use machine learning techniqes to address a prediction problem on __California Air Quality Index (AQI) (MAY CHANGE LATER)__. We want to investigate what factors that inflences AQI and use these factors to build a model to predict future AQI based on data that can be collected in the past.

__What model do you plan on using and why?__

AQI is an indicater index of air pollution level. [1] Air quality at any given time is an cummulative result of factors that's taken place in the past, including but not limited to previous air quality, climate and past air pollution. Due to the cummulative nature of our goal parameter, we want to choose a model that can iteratively take into acount of the information of the past period of time. Thus we choose to use sliding window approach to carve out data from a chosed period of time, and implement several models including multiple linear regression, symbolic regression, and support vector regression to irretatively make prediction on future AQI.

## Temperature & Precipitation

Using what we learned from analyzing this single dataset, we define a pipeline function to process each year's result in a weekly fashion.

```python
def temp_prcp_pipeline(start_yr, end_yr):
    df_list = []
    for yr in range(start_yr,end_yr+1):
        df = pd.read_csv("../data/daily_global_weather_{year}.csv".format(year = yr))
        df_LA_area = df[(df['Longitude'] <= -117.20715) &
                    (df['Longitude'] >= -119.20715) &
                    (df['Latitude'] <= 35.05316) &
                    (df['Latitude'] >= 33.05316)]
        
        # proper format
        df_LA_area.loc[:, "TEMP_C"] = df_LA_area.loc[:, "TAVG"] / 10.0
        df_LA_area.loc[:, "PRCP_MM"] = df_LA_area.loc[:, "PRCP"] / 10.0

        # format datetime
        df_LA_area["Datetime"] = pd.to_datetime(df_LA_area["Date"], format="%Y-%m-%d")
        df_LA_area["year"] = df_LA_area["Datetime"].dt.year
        df_LA_area["month"] = df_LA_area["Datetime"].dt.month
        df_LA_area["week"] = df_LA_area["Datetime"].dt.week
        df_LA_area["day"] = df_LA_area["Datetime"].dt.dayofyear

        df_LA_area["geo_cord"] = list(df_LA_area[['Latitude', 'Longitude']].itertuples(index=False, name=None))
        # use rg to search
        res = rg.search(df_LA_area["geo_cord"].to_list())

        states, county = [], []
        for i in res:
            states.append(i["admin1"])
            county.append(i["admin2"])

        # keep track of the information
        df_LA_area["state"] = states
        df_LA_area["county"] = county

        df_LA_area = df_LA_area[df_LA_area["state"] == 'California']

        df_LA_area = df_LA_area[["TEMP_C", "year", "month", "week", "day", "PRCP_MM"]]

        # aggregate
        df_LA_area = df_LA_area.groupby(["day"]).agg({
            "year": 'first',
            "month": 'first',
            "TEMP_C": np.mean,
            "PRCP_MM": np.mean,
        })

        df_LA_area = df_LA_area.reset_index()
        df_list.append(df_LA_area)
    # We concatenate the dataframes.
    concat_df = pd.concat(df_list)
    concat_df = concat_df.reset_index()

    return concat_df
```

## Pollutants

```python
def pollutants_pipeline(start_yr, end_yr):
    df_list = []
    for yr in range(start_yr,end_yr+1):
        co_df = pd.read_csv("../data/CA_daily_CO_{year}.csv".format(year = yr))
        no2_df = pd.read_csv("../data/CA_daily_NO2_{year}.csv".format(year = yr))
        so2_df = pd.read_csv("../data/CA_daily_SO2_{year}.csv".format(year = yr))

        # carve out LA area
        co_df = co_df[(no2_df['SITE_LONGITUDE'] <= -117.20715) & (no2_df['SITE_LONGITUDE'] >= -119.20715) &
                      (no2_df['SITE_LATITUDE'] <= 35.05316) & (no2_df['SITE_LATITUDE'] >= 33.05316)]
        no2_df = no2_df[(no2_df['SITE_LONGITUDE'] <= -117.20715) & (no2_df['SITE_LONGITUDE'] >= -119.20715) &
                        (no2_df['SITE_LATITUDE'] <= 35.05316) & (no2_df['SITE_LATITUDE'] >= 33.05316)]
        so2_df = so2_df[(so2_df['SITE_LONGITUDE'] <= -117.20715) & (so2_df['SITE_LONGITUDE'] >= -119.20715) &
                        (so2_df['SITE_LATITUDE'] <= 35.05316) & (so2_df['SITE_LATITUDE'] >= 33.05316)]

        # merging information
        all_pollutants = pd.merge(pd.merge(co_df, 
                                       no2_df, 
                                       how="outer", 
                                       left_on=['Date', 'Site ID'], 
                                       right_on=['Date', 'Site ID']), 
                              so2_df, 
                              how="outer", 
                              left_on=['Date', 'Site ID'], 
                              right_on=['Date', 'Site ID'])

        # keeping information
        all_pollutants_res = all_pollutants[['Date', 'Site ID', 
                                             'Daily Max 8-hour CO Concentration',
                                             'Daily Max 1-hour NO2 Concentration',
                                             'Daily Max 1-hour SO2 Concentration']]

        # adding more detailed time granularity
        all_pollutants_res["Datetime"] = pd.to_datetime(all_pollutants_res["Date"], format="%m/%d/%Y")
        all_pollutants_res["year"] = all_pollutants_res["Datetime"].dt.year
        all_pollutants_res["month"] = all_pollutants_res["Datetime"].dt.month
        all_pollutants_res["week"] = all_pollutants_res["Datetime"].dt.week
        all_pollutants_res["day"] = all_pollutants_res["Datetime"].dt.dayofyear 
        all_pollutants_res = all_pollutants_res.drop(columns=["Date"])

        # collect weekly info
        pollutants_by_week = all_pollutants_res.groupby('day').agg(np.mean)[[
            'Daily Max 8-hour CO Concentration', 
            'Daily Max 1-hour NO2 Concentration', 
            'Daily Max 1-hour SO2 Concentration'
        ]]

        pollutants_by_week = pollutants_by_week.rename(columns={
            'Daily Max 8-hour CO Concentration': 'avg max CO',
            'Daily Max 1-hour NO2 Concentration': "avg max NO2",
            'Daily Max 1-hour SO2 Concentration': 'avg max SO2'
        })

        pollutants_by_week = pollutants_by_week.reset_index()
        df_list.append(pollutants_by_week)
    # We concatenate the dataframes.
    concat_df = pd.concat(df_list)
    concat_df = concat_df.reset_index().iloc[:-71, :]  # collect till Oct 2020

    return concat_df
```

## Wind

```python
# helper function to change wind direction into categorial data
def apply_directions(df):
    df.loc[(df["WDF2"] >= 225) & (df["WDF2"] < 315), "W"] = 1
    df.loc[(df["WDF2"] >= 45) & (df["WDF2"] < 135), "E"] = 1
    df.loc[(df["WDF2"] >= 315) | (df["WDF2"] < 45), "N"] = 1
    df.loc[(df["WDF2"] >= 135) & (df["WDF2"] < 225), "S"] = 1
    df = df.drop(columns=["WDF2"])
    
    return df

def wind_initial_pipeline(start_yr, end_yr):
    df_list = []
    for yr in range(start_yr,end_yr+1):
        month_1_4 = pd.read_csv("../data/CA_daily_wind_{year}_1_4.csv".format(year = yr))
        month_5_8 = pd.read_csv("../data/CA_daily_wind_{year}_5_8.csv".format(year = yr))
        month_9_12 = pd.read_csv("../data/CA_daily_wind_{year}_9_12.csv".format(year = yr))      
        daily_wind = pd.concat([month_1_4, month_5_8, month_9_12])

        # carve out LA area
        daily_wind = daily_wind[(daily_wind['LONGITUDE'] <= -117.20715) & 
                                (daily_wind['LONGITUDE'] >= -119.20715) &
                                (daily_wind['LATITUDE'] <= 35.05316) & 
                                (daily_wind['LATITUDE'] >= 33.05316)]
        daily_wind = daily_wind.dropna()
        daily_wind['DATE'] = pd.to_datetime(daily_wind['DATE'])
        daily_wind["week"] = daily_wind["DATE"].dt.week
        daily_wind["year"] = daily_wind["DATE"].dt.year
        daily_wind["day"] = daily_wind["DATE"].dt.dayofyear
        daily_wind = daily_wind[['day','AWND', 'year', 'WDF2']].groupby('day').mean()
        daily_wind = daily_wind.rename(columns = {'AWND' : 'weekly_wind_speed'})
        daily_wind = apply_directions(daily_wind)
        daily_wind.fillna(0, inplace=True)
        daily_wind = daily_wind.reset_index()
        df_list.append(daily_wind)
        
    # We concatenate the dataframes.
    concat_df = pd.concat(df_list)
    concat_df = concat_df.reset_index().iloc[:-71, :]  # collect till Oct 2020
    return concat_df

```

# AQI nearby region

```python
def nearby_aqi_pipeline(start_yr, end_yr):
    E_df_list = []
    S_df_list = []
    W_df_list = []
    N_df_list = []
    for yr in range(start_yr,end_yr+1):
        df = pd.read_csv("../data/ad_viz_plotval_data_aqi_{year}.csv".format(year = yr))
            
        df["Datetime"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
        filtered_df = df[df["CBSA_CODE"].isnull()]
        df = df[~df.index.isin(filtered_df.index)]
        df["year"] = df["Datetime"].dt.year
        df["month"] = df["Datetime"].dt.month
        df["day"] = df["Datetime"].dt.dayofyear
        
        df = df[["year", "month", "day", "DAILY_AQI_VALUE", 'SITE_LONGITUDE', 'SITE_LATITUDE']]
        
                         
        E_df = df[(df['SITE_LONGITUDE'] <= -115.20715) & (df['SITE_LONGITUDE'] >= -117.20715) &
                (df['SITE_LATITUDE'] <= 35.55316) & (df['SITE_LATITUDE'] >= 32.55316)]
        S_df = df[(df['SITE_LONGITUDE'] <= -116.20715) & (df['SITE_LONGITUDE'] >= -120.20715) &
                (df['SITE_LATITUDE'] <= 33.05316) & (df['SITE_LATITUDE'] >= 31.05316)]
        W_df = df[(df['SITE_LONGITUDE'] <= -119.20715) & (df['SITE_LONGITUDE'] >= -121.20715) &
                (df['SITE_LATITUDE'] <= 35.55316) & (df['SITE_LATITUDE'] >= 32.55316)]
        N_df = df[(df['SITE_LONGITUDE'] <= -116.20715) & (df['SITE_LONGITUDE'] >= -120.20715) &
                (df['SITE_LATITUDE'] <= 37.05316) & (df['SITE_LATITUDE'] >= 35.05316)]
        

        # aggregate
        E_df = E_df.groupby(["day"]).agg({ "year": 'first', "month": 'first', "DAILY_AQI_VALUE": np.mean })
        S_df = S_df.groupby(["day"]).agg({ "year": 'first', "month": 'first', "DAILY_AQI_VALUE": np.mean })
        W_df = W_df.groupby(["day"]).agg({ "year": 'first', "month": 'first', "DAILY_AQI_VALUE": np.mean })
        N_df = N_df.groupby(["day"]).agg({ "year": 'first', "month": 'first', "DAILY_AQI_VALUE": np.mean })

        E_df = E_df.reset_index()
        E_df = E_df.rename(columns={"DAILY_AQI_VALUE":"E daily AQI"})
        E_df_list.append(E_df)
        
        S_df = S_df.reset_index()
        S_df = S_df.rename(columns={"DAILY_AQI_VALUE":"S daily AQI"})
        S_df_list.append(S_df)
        
        W_df = W_df.reset_index()
        W_df = W_df.rename(columns={"DAILY_AQI_VALUE":"W daily AQI"})
        W_df_list.append(W_df)
        
        N_df = N_df.reset_index()
        N_df = N_df.rename(columns={"DAILY_AQI_VALUE":"N daily AQI"})
        N_df_list.append(N_df)
        
    # We concatenate the dataframes.
    concat_E_df = pd.concat(E_df_list)
    concat_E_df = concat_E_df.reset_index().iloc[:-71, :]  # collect till Oct 2020
    
    # We concatenate the dataframes.
    concat_S_df = pd.concat(S_df_list)
    concat_S_df = concat_S_df.reset_index().iloc[:-71, :]  # collect till Oct 2020
    
    # We concatenate the dataframes.
    concat_W_df = pd.concat(W_df_list)
    concat_W_df = concat_W_df.reset_index().iloc[:-71, :]  # collect till Oct 2020
    
    # We concatenate the dataframes.
    concat_N_df = pd.concat(N_df_list)
    concat_N_df = concat_N_df.reset_index().iloc[:-71, :]  # collect till Oct 2020
    
    
    # After concating the dataframes, we calculate delta aqi of nearby area and area of interest/
    # That is, the changing value of the average aqi from current week to previous week
    merge_df_1 = pd.merge(concat_E_df, concat_S_df, left_index=True, right_index=True)
    merge_df_2 = pd.merge(concat_W_df, concat_N_df, left_index=True, right_index=True)
    merge_df = pd.merge(merge_df_1, merge_df_2, left_index=True, right_index=True)
    return merge_df[['E daily AQI', 'S daily AQI', 'W daily AQI', 'N daily AQI']]

```

## AQI

```python
def aqi_pipeline(start_yr, end_yr):
    df_list = []
    for yr in range(start_yr,end_yr+1):
        df = pd.read_csv("../data/ad_viz_plotval_data_aqi_{year}.csv".format(year = yr))        
        df = df[(df['SITE_LONGITUDE'] <= -117.20715) & (df['SITE_LONGITUDE'] >= -119.20715) &
                (df['SITE_LATITUDE'] <= 35.05316) & (df['SITE_LATITUDE'] >= 33.05316)]
        df = df[["Date", "Site ID", "Daily Mean PM2.5 Concentration", "DAILY_AQI_VALUE", "CBSA_CODE", "STATE"]]
        df["Datetime"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
        filtered_df = df[df["CBSA_CODE"].isnull()]
        df = df[~df.index.isin(filtered_df.index)]
        df["year"] = df["Datetime"].dt.year
        df["month"] = df["Datetime"].dt.month
        df["week"] = df["Datetime"].dt.week
        df["day"] = df["Datetime"].dt.dayofyear

        df = df[["year", "month", "week", "day", "DAILY_AQI_VALUE", "STATE", "Site ID"]]
        df = df[df["STATE"] == "California"]

        # aggregate
        df = df.groupby(["day"]).agg({
            "year": 'first',
            "month": 'first',
            "DAILY_AQI_VALUE": np.mean
        })

        df = df.reset_index()
        df = df.rename(columns={"DAILY_AQI_VALUE":"weekly AQI"})
        df_list.append(df)
        
    # We concatenate the dataframes.
    concat_df = pd.concat(df_list)
    concat_df = concat_df.reset_index().iloc[:-71, :]  # collect till Oct 2020
    # After concating the dataframes, we calculate delta aqi/
    # That is, the changing value of the average aqi from current week to previous week
    concat_df['weekly delta AQI'] = 0
    concat_df['weekly delta AQI'][1:] = np.array(concat_df['weekly AQI'][1:]) - np.array(concat_df['weekly AQI'][:-1])
    return concat_df
```

## Constructing the final dataframe

```python
def pipeline(start_yr, end_yr):
    # temp, prcp
    ca_tp = temp_prcp_pipeline(start_yr, end_yr)
    df = ca_tp

    # wind_speed
    ca_wind = wind_pipeline(start_yr, end_yr)
    df['weekly wind speed'] = ca_wind['weekly_wind_speed']
    df['W'] = ca_wind['W']
    df['S'] = ca_wind['S']
    df['E'] = ca_wind['E']

    # co, no2, so2
    ca_pollu = pollutants_pipeline(start_yr, end_yr)
    df["weekly CO"] = ca_pollu["avg max CO"]
    df["weekly NO2"] = ca_pollu["avg max NO2"]
    df["weekly SO2"] = ca_pollu["avg max SO2"]

    # aqi
    ca_aqi = aqi_pipeline(start_yr, end_yr)
    df["weekly AQI"] = ca_aqi["weekly AQI"]
    df["weekly delta AQI"] = ca_aqi["weekly delta AQI"]
    
    # aqi of nearby region
    nearby_aqi = nearby_aqi_pipeline(start_yr, end_yr)
    df['E delta aqi'] = (nearby_aqi['E daily AQI'] - df["weekly AQI"]) * df['E'] * df['weekly wind speed']
    df['S delta aqi'] = nearby_aqi['S daily AQI'] - df["weekly AQI"] * df['S'] * df['weekly wind speed']
    df['W delta aqi'] = nearby_aqi['W daily AQI'] - df["weekly AQI"] * df['W'] * df['weekly wind speed']
    df['N delta aqi'] = nearby_aqi['N daily AQI'] - df["weekly AQI"] * df['weekly wind speed']

    df = df[["day", "year", "month", "TEMP_C", "PRCP_MM", 
             "weekly CO", "weekly NO2", "weekly SO2", 
             'weekly wind speed','W', "S", 'E', 
             'E delta aqi', 'S delta aqi', 'W delta aqi', 'N delta aqi', 
             "weekly delta AQI"]]
    df.insert(loc=0, column="time_delta", value=df.index)
    return df

final_df = pipeline(2016, 2020)
final_df.head()
```

We want to normalize the input features

```python
final_df_normalized = (final_df.iloc[:, :-1] - np.mean(final_df.iloc[:, :-1], axis=0)) / np.std(final_df.iloc[:, :-1], axis=0)

# retrieve the ys
final_df_normalized_with_y = final_df_normalized.copy()
final_df_normalized_with_y["weekly delta AQI"] = final_df["weekly delta AQI"]

final_df_normalized_with_y = final_df_normalized_with_y[["time_delta", "day", "year", "month", 
                                                         "TEMP_C", "PRCP_MM", 
                                                         "weekly CO", "weekly NO2", "weekly SO2", 
                                                         "weekly wind speed", "W", "S", 'E',
                                                         'E delta aqi', 'S delta aqi', 
                                                         'W delta aqi', 'N delta aqi',
                                                          "weekly delta AQI"]]

final_df_normalized_with_y
```

## Modeling

In this section, we define the modeling process and establish the metric to evaluate the prediction results.

## Window Slider

```python
# referenced from: https://towardsdatascience.com/ml-approaches-for-time-series-4d44722e48fe
class WindowSlider(object):
    
    def __init__(self, window_size = 5):        
        '''
        Window Slider object
        ====================
        w: window_size - number of time steps to look back
        o: offset between last reading and temperature
        r: response_size - number of time steps to predict
        l: maximum length to slide - (#observation - w)
        p: final predictors - (#predictors * w)
        '''
        self.w = window_size
        self.o = 0
        self.r = 1       
        self.l = 0
        self.p = 0
        self.names = []
        
    def re_init(self, arr):
        '''
        Helper function to initializate to 0 a vector
        '''
        arr = np.cumsum(arr)
        return arr - arr[0]
                

    def collect_windows(self, X, window_size=5, offset=0, previous_y=False):
        '''
        Input: X is the input matrix, each column is a variable
        Returns: diferent mappings window-output
        '''
        cols = len(list(X)) - 1
        N = len(X)
        
        self.o = offset
        self.w = window_size
        self.l = N - (self.w + self.r) + 1
        if not previous_y: self.p = cols * (self.w)
        if previous_y: self.p = (cols + 1) * (self.w)
        
        # Create the names of the variables in the window
        # Check first if we need to create that for the response itself
        if previous_y: x = copy.deepcopy(X)
        if not previous_y: x = X.drop(X.columns[-1], axis=1)  
        
        for j, col in enumerate(list(x)):        
                
            for i in range(self.w):
                
                name = col + ('(%d)' % (i+1))
                self.names.append(name)
        
        # Incorporate the timestamps where we want to predict
        for k in range(self.r):
            
            name = '∆t' + ('(%d)' % (self.w + k + 1))
            self.names.append(name)
            
        self.names.append('Y')
                
        df = pd.DataFrame(np.zeros(shape=(self.l, (self.p + self.r + 1))), 
                          columns=self.names)
        
        # Populate by rows in the new dataframe
        for i in range(self.l):
            
            slices = np.array([])
            
            # Flatten the lags of predictors
            for p in range(x.shape[1]):
            
                line = X.values[i:self.w + i, p]
                # Reinitialization at every window for ∆T
                if p == 0: line = self.re_init(line)
                    
                # Concatenate the lines in one slice    
                slices = np.concatenate((slices, line)) 
 
            # Incorporate the timestamps where we want to predict
            line = np.array([self.re_init(X.values[i:i+self.w+self.r, 0])[-1]])
            y = np.array(X.values[self.w + i + self.r - 1, -1]).reshape(1,)
            slices = np.concatenate((slices, line, y))
            
            # Incorporate the slice to the cake (df)
            df.iloc[i,:] = slices
            
        return df
```

## Utilities

```python
def evaluate_rmse(y, y_pred, model_name, status, suppress=False):
    # sanity check
    assert len(y) == len(y_pred)
    rmse = np.sqrt(np.sum(np.power(y_pred - y,2)) / len(y_pred - y))
    if not suppress:
        print('%s RMSE of %s = %.2f' % (status, model_name, rmse))
    return rmse
```

```python
def evaluate_mae(y, y_pred, model_name, status, suppress=False):
    # sanity check
    assert len(y) == len(y_pred)
    mae = np.sum(np.abs(y_pred - y))
    if not suppress:
        print('%s MAE of %s = %.2f' % (status, model_name, mae))
    return mae
```

__Q: Does your model require hyperparameter tuning?__

Yes. As indicated below, we have proposed 4 models (Multiple linear regression, Symbolic regression, Multiple layer perceptron, and Random forest regression). For each model, we consider the case of using polynomial lifting map on our features or not. Then, it is evident that each model requires different level of hyperparameter tuning. For every model, we will tune the sliding window size and whether to include the previous week's AQI as features into the model. Then, depending on the model, we have different model-dependent parameters to tune:

For linear regression with l2 penality, we tune the alpha parameter.
For symbolic regreesion, we tune for the number of generations and population size.
For multiple layer perceptron, we tune for the alpha (l2 penalty), and the hidden layer dimensions.
For random forest regression, we tune for the number of trees in forest, the proportion of samples to be used in each tree, and the proportion of features to be used in each split.

__Q: How do you perform cross validation on your model?__

For every model, we define a corresponding cross validation function as shown in the below sections. The cross validation function takes in several lists, each of which contain the parameters to be tuned on that model. We always tune on the training set, spliting it k folds, using every k-1 folds for training the model on that set of parameters and then use the remaining 1 fold for validation. We repreat this process k times for each set of parameters. Then, we take the average as the final validation score. Because we are handling a regression problem, we use rmse as the evaluation metric. The lower the score, the better fit in our case. Finally, we use the best tuned parameters to construct the final model, train on the training set and evaluate on the test set. Note that, during cross validation, the model never gets to see the test set.

__Q: What loss metrics are you using to evaluate your model?__

As indicated below and by some preliminary experiments, we are using root mean squared error as the loss metric. Since this is a regression task and we are fitting to the AQI curve, a good fitting curve would produce a small rmse and we want to penalize bad predictions that are way off from the curve.

__Q: From a bias-variance tradeoff standpoint, how do you assess the performance of your model? How do you check if it is overfitting?__

In general, based on the EDA and the visulization of the training data (as shown below in later sections), we see oscillation and some level of noise in the data. So, we don't want our models to overfit to the noise in the training set, and we control that with the use of l2 penality alpha. Empircally from our experiments, we believe the existing features we engineered could get a good fit with a rmse near 12. Since we have not introduced any new features, we set our goal to minimize the test rmse to that standard.

For the bias-variance tradeoff, we can use the best tuned model and run it several times to evaluate the variance and bias. It is easy to check overfitting. We have included helpful plots in the following sections that allow direct visualization on the predicted y and the actual y for both training and test sets. It the model is overfitting, we would see a completely overlapped curve for training, but the test graph could not grasp the trend.

__Q: How would you improve your model based on the outcome?__

We use cross validation to find the best performing model parameters. To improve the model, we consider two things:

1. Use a polynomial lifting map to lift the features into a high dimensional space that contains non-linearity. As explained previously, we introduce polynomial features to grant more power for the models to model the oscillation and non-linearity in the AQI trend.
2. Consider the previous AQI values as an additional features. This is based on the discovery that AQI values are time-related, i.e., the next day's AQI will very likely to deviate not much from the past few day's AQI. Though this may seem to be cheating, we consider this as an additional feature for the purpose of exploring a better fit.
3. Use more data. Previously we used 2018-2019's data as training and predict on 2020. Now we are using 2016-2019's data and predict on 2020. We see a reduction in rmse on all models, as the amount of training examples increased (to allow the model learn more about the seasons, for example).

## Multiple Linear Regression (including polynomial features)

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures

class MultipleLinearRegression:
    
    def __init__(self, window_size, tn_data, vd_data, with_y_inc=False, 
                 regularization=None, alpha=2.0, poly_feature=False, poly_degree=3):
        self.window_size = window_size
        self.with_y_inc = with_y_inc
        self.regularization = regularization
        
        # construct the processed data from sliding window
        if self.with_y_inc:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
        else:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)   

        if poly_feature:
            polynomial_features = PolynomialFeatures(degree=poly_degree)
            tn_poly_X = polynomial_features.fit_transform(self.tn_data.iloc[:,:-1])

            self.tn_data_X = tn_poly_X
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            vd_poly_X = polynomial_features.transform(self.vd_data.iloc[:,:-1])
            self.vd_data_X = vd_poly_X
            self.vd_data_Y = self.vd_data.iloc[:,-1] 
        else:
            # extract the X, Y from the data
            self.tn_data_X = self.tn_data.iloc[:,:-1]
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            self.vd_data_X = self.vd_data.iloc[:,:-1]
            self.vd_data_Y = self.vd_data.iloc[:,-1]        
        
        if self.regularization is None:
            self.model = LinearRegression(fit_intercept=True)
        elif self.regularization == "ridge":
            self.model = Ridge(fit_intercept=True, alpha=alpha)
        elif self.regularization == "lasso":
            self.model = Lasso(fit_intercept=True, alpha=alpha)
        else:
            raise Exception("The model type is badly defined.")
        
    def train(self): 
        self.model.fit(self.tn_data_X, self.tn_data_Y)
    
    def get_XY(self):
        return self.tn_data, self.tn_data_X, self.tn_data_Y, self.vd_data, self.vd_data_X, self.vd_data_Y
    
    def predict(self, X):
        return self.model.predict(X)
```

```python
def linear_regression_k_fold_cv(shuffled_data, shuffled_labels, k, window_list, 
                                poly_feature=False, poly_degree=2):
    total_num_samples = shuffled_data.shape[0]
    # score holder
    scores_vd = list()
    my_range = int(total_num_samples // k)

    # keep track of the best info
    best_window = 1
    best_score = np.inf

    for window in window_list:
        score = 0.0
        # k fold partition on the indexes, repeat k times
        for i in range(k):
            # boundary checking
            end = (i+1)*my_range if (i+1)*my_range <= total_num_samples else total_num_samples
            # specify the kth validation set
            vd_data = shuffled_data.iloc[i*my_range:end, :]
            vd_labels = shuffled_labels.iloc[i*my_range:end]
            vd_final = pd.concat([vd_data, vd_labels], axis=1)

            # specify the k-1 training set (using concatenation)
            tn_data = pd.concat([shuffled_data.iloc[0:i*my_range, :], 
                                      shuffled_data.iloc[end:total_num_samples, :]])
            tn_labels = pd.concat([shuffled_labels.iloc[0:i*my_range], 
                                        shuffled_labels.iloc[end:total_num_samples]])
            tn_final = pd.concat([tn_data, tn_labels], axis=1)

            # define the model
            model = MultipleLinearRegression(window, tn_final, vd_final, with_y_inc=False, 
                                             regularization=None, alpha=None, 
                                             poly_feature=poly_feature, poly_degree=poly_degree)

            # extract the data
            _, _, _, _, vd_data_X, vd_data_Y = model.get_XY()

            # train the model!!! ：）
            model.train()

            # get the rmse
            vd_pred = model.predict(vd_data_X)
            vd_acy = evaluate_rmse(vd_data_Y, vd_pred, "multiple linear regression", 
                                   "validation_" + str(i), suppress=True)
            score += vd_acy

        score = 1.0 * score / k

        if score < best_score:
            best_score = score
            best_window = window

        scores_vd.append(score)

        # print the result
        print("window %d, score %f" % (window, score))

        # plotter (later)
    return best_window, best_score
```

```python
def linear_regression_ridge_k_fold_cv(shuffled_data, shuffled_labels, k, window_list, alpha_list,
                                      poly_feature=False, poly_degree=2):
    total_num_samples = shuffled_data.shape[0]
    # score holder
    scores_vd = list()
    my_range = int(total_num_samples // k)

    # keep track of the best info
    best_alpha = 0.0
    best_window = 1
    best_score = np.inf

    for alpha in alpha_list:
        for window in window_list:
            score = 0.0
            # k fold partition on the indexes, repeat k times
            for i in range(k):
                # boundary checking
                end = (i+1)*my_range if (i+1)*my_range <= total_num_samples else total_num_samples
                # specify the kth validation set
                vd_data = shuffled_data.iloc[i*my_range:end, :]
                vd_labels = shuffled_labels.iloc[i*my_range:end]
                vd_final = pd.concat([vd_data, vd_labels], axis=1)

                # specify the k-1 training set (using concatenation)
                tn_data = pd.concat([shuffled_data.iloc[0:i*my_range, :], 
                                          shuffled_data.iloc[end:total_num_samples, :]])
                tn_labels = pd.concat([shuffled_labels.iloc[0:i*my_range], 
                                            shuffled_labels.iloc[end:total_num_samples]])
                tn_final = pd.concat([tn_data, tn_labels], axis=1)
                        
                # define the model
                model = MultipleLinearRegression(window, tn_final, vd_final, with_y_inc=False, 
                                                 regularization="ridge", alpha=alpha, 
                                                 poly_feature=poly_feature, poly_degree=poly_degree)
                
                # extract the data
                _, _, _, _, vd_data_X, vd_data_Y = model.get_XY()
                
                # train the model!!!
                model.train()
                
                # get the rmse
                vd_pred = model.predict(vd_data_X)
                vd_acy = evaluate_rmse(vd_data_Y, vd_pred, "multiple linear regression", 
                                       "validation_" + str(i), suppress=True)
                score += vd_acy

            score = 1.0 * score / k

            if score < best_score:
                best_score = score
                best_alpha = alpha
                best_window = window

            scores_vd.append(score)

            # print the result
            print("window %d, alpha %f, score %f" % (window, alpha, score))

        # plotter (later)
    return best_window, best_alpha, best_score
```

## Symbolic regressor

```python
import gplearn as gpl        
from gplearn.genetic import SymbolicRegressor

# reference: https://gplearn.readthedocs.io/en/stable/reference.html#gplearn.genetic.SymbolicRegressor
class SymbolicRegression:
    
    def __init__(self, window_size, tn_data, vd_data, function_set, with_y_inc=False, 
                 population_size=5000, tournament_size=5, generations=40, p_crossover=0.7,
                 p_subtree_mutation=0.1, p_hoist_mutation=0.05, p_point_mutation=0.1,
                 max_samples=0.9, parsimony_coeff=0.01, verbose=1, 
                 random_state=1, poly_feature=False, poly_degree=2):
        self.window_size = window_size
        self.with_y_inc = with_y_inc
        self.function_set = function_set
        self.population_size = population_size
        self.tournament_size = tournament_size
        self.generations = generations
        self.p_crossover = p_crossover
        self.p_subtree_mutation = p_subtree_mutation
        self.p_hoist_mutation = p_hoist_mutation
        self.p_point_mutation = p_point_mutation
        self.verbose = verbose
        self.random_state = random_state
        self.max_samples = max_samples
        self.parsimony_coeff = parsimony_coeff
        
        # construct the processed data from sliding window
        if self.with_y_inc:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
        else:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)   

        if poly_feature:
            polynomial_features = PolynomialFeatures(degree=poly_degree)
            tn_poly_X = polynomial_features.fit_transform(self.tn_data.iloc[:,:-1])

            self.tn_data_X = tn_poly_X
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            vd_poly_X = polynomial_features.transform(self.vd_data.iloc[:,:-1])
            self.vd_data_X = vd_poly_X
            self.vd_data_Y = self.vd_data.iloc[:,-1] 
        else:
            # extract the X, Y from the data
            self.tn_data_X = self.tn_data.iloc[:,:-1]
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            self.vd_data_X = self.vd_data.iloc[:,:-1]
            self.vd_data_Y = self.vd_data.iloc[:,-1]          
        
        self.model = SymbolicRegressor(population_size=self.population_size, 
                                       tournament_size=self.tournament_size,
                                       generations=self.generations, stopping_criteria=0.1,
                                       function_set=self.function_set, metric='rmse',
                                       p_crossover=self.p_crossover, 
                                       p_subtree_mutation=self.p_subtree_mutation,
                                       p_hoist_mutation=self.p_hoist_mutation, 
                                       p_point_mutation=self.p_point_mutation,
                                       verbose=self.verbose, 
                                       random_state=self.random_state, n_jobs=-1,
                                       parsimony_coefficient=self.parsimony_coeff,
                                       max_samples=self.max_samples)
        
    def train(self): 
        self.model.fit(self.tn_data_X, self.tn_data_Y)
    
    def get_XY(self):
        return self.tn_data, self.tn_data_X, self.tn_data_Y, self.vd_data, self.vd_data_X, self.vd_data_Y
    
    def predict(self, X):
        return self.model.predict(X)
```

## Multiple Layer Perceptron (MLP)

```python
from sklearn.neural_network import MLPRegressor

class MultipleLayerPerceptron:
    
    def __init__(self, window_size, tn_data, vd_data, with_y_inc=False,
                 max_iter=4000, random_state=1, hidden_layer_sizes=(200,100),
                 shuffle=True, alpha=0.01, batch_size=200, poly_feature=False, poly_degree=2):
        self.window_size = window_size
        self.with_y_inc = with_y_inc
        self.max_iter = max_iter
        self.random_state = random_state
        self.hidden_layer_sizes = hidden_layer_sizes
        self.shuffle = shuffle
        self.alpha = alpha
        self.batch_size = batch_size

        # construct the processed data from sliding window
        if self.with_y_inc:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
        else:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)   

        if poly_feature:
            polynomial_features = PolynomialFeatures(degree=poly_degree)
            tn_poly_X = polynomial_features.fit_transform(self.tn_data.iloc[:,:-1])

            self.tn_data_X = tn_poly_X
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            vd_poly_X = polynomial_features.transform(self.vd_data.iloc[:,:-1])
            self.vd_data_X = vd_poly_X
            self.vd_data_Y = self.vd_data.iloc[:,-1] 
        else:
            # extract the X, Y from the data
            self.tn_data_X = self.tn_data.iloc[:,:-1]
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            self.vd_data_X = self.vd_data.iloc[:,:-1]
            self.vd_data_Y = self.vd_data.iloc[:,-1]        
        
        self.model = MLPRegressor(random_state=self.random_state, max_iter=self.max_iter,
                                  hidden_layer_sizes=self.hidden_layer_sizes, shuffle=self.shuffle,
                                  alpha=self.alpha, batch_size=self.batch_size)
        
    def train(self): 
        self.model.fit(self.tn_data_X, self.tn_data_Y)
    
    def get_XY(self):
        return self.tn_data, self.tn_data_X, self.tn_data_Y, self.vd_data, self.vd_data_X, self.vd_data_Y
    
    def predict(self, X):
        return self.model.predict(X)
```

```python
def mlp_k_fold_cv(shuffled_data, shuffled_labels, k, window_list, alpha_list, hidden_layer_list,
                  poly_feature=False, poly_degree=2, with_y_inc=False):
    total_num_samples = shuffled_data.shape[0]
    # score holder
    scores_vd = list()
    my_range = int(total_num_samples // k)

    # keep track of the best info
    best_alpha = 0.0
    best_window = 1
    best_hl = (100,)
    best_score = np.inf

    for hl in hidden_layer_list:
        for alpha in alpha_list:
            for window in window_list:
                score = 0.0
                # k fold partition on the indexes, repeat k times
                for i in range(k):
                    # boundary checking
                    end = (i+1)*my_range if (i+1)*my_range <= total_num_samples else total_num_samples
                    # specify the kth validation set
                    vd_data = shuffled_data.iloc[i*my_range:end, :]
                    vd_labels = shuffled_labels.iloc[i*my_range:end]
                    vd_final = pd.concat([vd_data, vd_labels], axis=1)

                    # specify the k-1 training set (using concatenation)
                    tn_data = pd.concat([shuffled_data.iloc[0:i*my_range, :], 
                                              shuffled_data.iloc[end:total_num_samples, :]])
                    tn_labels = pd.concat([shuffled_labels.iloc[0:i*my_range], 
                                                shuffled_labels.iloc[end:total_num_samples]])
                    tn_final = pd.concat([tn_data, tn_labels], axis=1)

                    # define the model
                    model = MultipleLayerPerceptron(window, tn_final, vd_final, with_y_inc=with_y_inc, 
                                                     hidden_layer_sizes=hl, alpha=alpha, 
                                                     poly_feature=poly_feature, poly_degree=poly_degree)

                    # extract the data
                    _, _, _, _, vd_data_X, vd_data_Y = model.get_XY()

                    # train the model!!!
                    model.train()

                    # get the rmse
                    vd_pred = model.predict(vd_data_X)
                    vd_acy = evaluate_rmse(vd_data_Y, vd_pred, "multiple layer perceptron", 
                                           "validation_" + str(i), suppress=True)
                    score += vd_acy

                score = 1.0 * score / k

                if score < best_score:
                    best_score = score
                    best_alpha = alpha
                    best_window = window
                    best_hl = hl

                scores_vd.append(score)

                # print the result
                print("window %d, alpha %f, hiddlen layer %s, score %f" % (window, alpha, hl, score))

        # plotter (later)
    return best_window, best_alpha, best_hl, best_score
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor

class RandomForestRegression:
    
    def __init__(self, window_size, tn_data, vd_data, with_y_inc=False,
                 n_estimators=100, max_depth=10, poly_feature=False, poly_degree=2,
                 max_samples=0.8, criterion='mse', max_features=0.8):
        self.window_size = window_size
        self.with_y_inc = with_y_inc
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.max_samples = max_samples
        self.criterion = criterion
        self.max_features = max_features

        # construct the processed data from sliding window
        if self.with_y_inc:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=True, 
                                                          window_size=self.window_size)
        else:
            self.tn_data = WindowSlider().collect_windows(tn_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)
            self.vd_data = WindowSlider().collect_windows(vd_data.iloc[:,0:], previous_y=False, 
                                                          window_size=self.window_size)   

        if poly_feature:
            polynomial_features = PolynomialFeatures(degree=poly_degree)
            tn_poly_X = polynomial_features.fit_transform(self.tn_data.iloc[:,:-1])

            self.tn_data_X = tn_poly_X
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            vd_poly_X = polynomial_features.transform(self.vd_data.iloc[:,:-1])
            self.vd_data_X = vd_poly_X
            self.vd_data_Y = self.vd_data.iloc[:,-1] 
        else:
            # extract the X, Y from the data
            self.tn_data_X = self.tn_data.iloc[:,:-1]
            self.tn_data_Y = self.tn_data.iloc[:,-1]
            self.vd_data_X = self.vd_data.iloc[:,:-1]
            self.vd_data_Y = self.vd_data.iloc[:,-1]         
        
        self.model = ExtraTreesRegressor(random_state=0, n_estimators=self.n_estimators,
                                         max_depth=self.max_depth, max_samples=self.max_samples, 
                                         criterion=self.criterion, n_jobs=-1, max_features=max_features)
        
    def train(self): 
        self.model.fit(self.tn_data_X, self.tn_data_Y)
    
    def get_XY(self):
        return self.tn_data, self.tn_data_X, self.tn_data_Y, self.vd_data, self.vd_data_X, self.vd_data_Y
    
    def predict(self, X):
        return self.model.predict(X)
```
